# 05. RNN 언어 모델(Recurrent Neural Network Language Model, RNNLM) 

1. RNN 언어 모델
- 입력의 길이 고정할 필요 없음 
- 교사 강요(teacher forcing): test 과정에서 t 시점의 출력이 t+1 시점의 입력으로 사용되는 RNN 모델을 훈련시킬 때 사용하는 훈련 기법 
- training의 출력층의 활성 함수는 softmax, 손실 함수는 cross-entropy 사용 
- Embedding 행렬 = 단어 집합의 크기 * 임베딩 벡터의 크기
